# Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Importing Dataset

```python
dataset=pd.read_csv("quality.csv")
```

```python
dataset
```

# Number of Rows & Column

```python
print("Number of rows : " , dataset.shape[0])
print("Number of columns :" , dataset.shape[1])
```

# Univariate Analysis

```python
import seaborn as sb
sb.distplot(dataset["MedicalClaims"],hist=True,kde=True,kde_kws={'color': 'red'})
```

# Bivariate Analysis : Distribution Plot

```python
sb.jointplot(x="MedicalClaims",y="AcuteDrugGapSmall",data=dataset)
plt.show()
```

```python
sb.jointplot(x="MedicalClaims",y="AcuteDrugGapSmall",data=dataset,kind='hex')
plt.show()
```

```python
sb.jointplot(x="MedicalClaims",y="AcuteDrugGapSmall",data=dataset,kind='kde')
plt.show()
```

# Bivariate Analysis : Pair Plot

```python
sb.pairplot(dataset,hue='AcuteDrugGapSmall',diag_kind='kde',kind='scatter',palette='husl')
plt.show()
```

# Catagorical Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,4))
sns.stripplot(x="MedicalClaims",y="PoorCare",data=dataset,jitter=True)
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,4))
sns.swarmplot(x="MedicalClaims",y="PoorCare",data=dataset)
plt.xticks(rotation=45)
plt.show()
```

# Distribution Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,4))
sns.violinplot(x="MedicalClaims",y="PoorCare",data=dataset)
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,4))
sns.boxplot(x="MedicalClaims",y="PoorCare",data=dataset,)
plt.xticks(rotation=45)
plt.show()
```

# Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,4))
sns.barplot(x="MedicalClaims",y="PoorCare",data=dataset)
plt.xticks(rotation=45)
plt.show()
```

# Point Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,4))
sns.pointplot(x="MedicalClaims",y="PoorCare",data=dataset,)
plt.xticks(rotation=45)
plt.show()
```
